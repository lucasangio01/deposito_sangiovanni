```python
import numpy as np
from lez2_1 import *
```

```python
path = "../data/Mall_customers.csv"
data = load_data(path)
```

```python
data.head()
```

```python
features = ['Annual Income (k$)', 'Spending Score (1-100)']
X = data[features]
```

```python
X_scaled = standardize(X)
```

```python
K, silhouette_scores, kmeans, labels = apply_kmeans(X_scaled)
```

```python
print(silhouette_scores)
print(f"\nAverage score: {round(np.mean(silhouette_scores), 3)}")
```

```python
plot_silhouette(K, silhouette_scores)
```

Il numero di cluster più adatto, seguendo il metodo "elbow", è 5. Procediamo a creare un K-means con k = 5.

```python
kmeans_5, labels_5 = new_kmeans(X_scaled)
```

```python
plot_datapoints_clusters(X_scaled, kmeans_5, labels_5)
```

I cluster più in alto a destra rappresentano i consumatori che spendono di più, e che hanno un reddito più alto. Al contrario, in basso a sinistra ci sono i consumatori che spendono di meno e che hanno un reddito più basso.
In basso a destra i consumatori che hanno un reddito più alto ma che spendono di meno ("tirchi"), e in alto a sinistra quelli che hanno un reddito più basso e spendono molto.
